Importing the dataset


```python
dataset = pd.read_csv('Students.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values 
print(x)
```

```python
print(y)
```

# Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder ='passthrough')
print(ct)
```

```python
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/10, random_state = 0 )
```

```python
x_test
```

# Training the multiple Linear Regression model on the training dataset


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

# Predicting the test results

```python
y_pred = regressor.predict(x_test) 
np.set_printoptions(precision = 2) 
print('\n\nPredicted Actual VS Test')
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```
